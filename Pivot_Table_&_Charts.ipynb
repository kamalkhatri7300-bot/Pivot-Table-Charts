Q4. Dxplain the difference between Sorting and Filtering with an example from this dataset.

- Answer 4

Sorting: Arranges the existing data in a specific order (e.g., A to Z or High to Low) without removing any rows.

Example: Sorting the Sales column from Largest to Smallest to see which transaction earned the most money (the ₹497 Oranges sale would move to the top).

Filtering: Hides data that does not meet specific criteria, showing only the rows you want to see.

Example: Filtering the Product column to show only "Apples". This would hide all Oranges and Mangoes rows, leaving only the 5 Apple transactions visible.

Q7. Why are Dashboards important for decision-making in business analysis?

- Answer 7


- Identify Trends: Quickly spot whether sales are increasing or decreasing over time.

- Monitor KPIs: Track key performance indicators like Total Sales and Quantity at a glance.

- Enable Interactivity: Use tools like Slicers to filter data instantly, helping to answer specific business questions (e.g., "How did West perform vs East?").
